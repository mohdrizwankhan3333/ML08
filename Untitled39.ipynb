```python
'''
### Q1: Difference Between Simple and Multiple Linear Regression

Simple Linear Regression: Involves one independent variable to predict a dependent variable.  
Example: Predicting salary based on years of experience.

Multiple Linear Regression: Involves multiple independent variables to predict a dependent variable.  
Example: Predicting house prices based on size, location, and number of bedrooms.

### Q2: Assumptions of Linear Regression and Checking Them

1. Linearity: Relationship between independent and dependent variables is linear.
2. Independence: Observations are independent.
3. Homoscedasticity: Constant variance of errors.
4. Normality: Errors are normally distributed.

Checking: Residual plots, Durbin-Watson test, Q-Q plots, and statistical tests.

### Q3: Interpreting Slope and Intercept in Linear Regression

Slope: Change in dependent variable for a one-unit change in independent variable.  
Intercept: Value of dependent variable when all independent variables are zero.

Example: Predicting height from age: Slope shows height increase per year, intercept is height at age zero.

### Q4: Concept of Gradient Descent

Gradient Descent: Optimization algorithm to minimize the cost function by iteratively updating parameters in 
the opposite direction of the gradient.
Use in Machine Learning: To find the optimal parameters of models like linear regression.

### Q5: Multiple Linear Regression Model

Multiple Linear Regression: Predicts a dependent variable using multiple independent variables.  
Difference: Unlike simple linear regression, it accounts for multiple factors influencing the outcome.

### Q6: Multicollinearity in Multiple Linear Regression

Multicollinearity: When independent variables are highly correlated, causing instability in coefficient 
estimates.

Detection: Variance Inflation Factor (VIF), correlation matrix.  
Addressing: Removing correlated predictors, combining variables, or regularization techniques.

### Q7: Polynomial Regression Model

Polynomial Regression: Fits a nonlinear relationship by transforming variables into polynomial terms.

Difference: Unlike linear regression, it can model curves by including higher-degree terms of the predictors.

### Q8: Advantages and Disadvantages of Polynomial Regression

Advantages: Can model complex, nonlinear relationships.

Disadvantages: Prone to overfitting, computationally expensive.

Preferred Use: When the relationship between variables is nonlinear and cannot be captured by linear 
regression.
```
